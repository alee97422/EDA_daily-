```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/bad-passwords-never-use-them/passwords.csv', encoding='latin1')
```

```python

words = df['password'].tolist()
```

```python
# Convert the list of words to a single string
# Convert the list of words to a single string, ensuring all items are strings
word_data = ' '.join(str(word) for word in words)
```

```python
# Create the WordCloud
wordcloud = WordCloud(width=800, height=800, background_color='black').generate(word_data)
```

```python
# Plot the WordCloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# Show the WordCloud (optional)
plt.show()
```
